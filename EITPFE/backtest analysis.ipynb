```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv("trades1.csv", parse_dates=True, index_col=0)
df.head(10)
```

```
dfa = df.dropna(subset=['curPerf'])
dfa.sort_index(inplace=True)
dfa

```

```
dfa.groupby(['Symbol']).Symbol.count()
```

```
print(dfa[dfa.curPerf>0].groupby(['Symbol']).Symbol.count())
print(dfa[dfa.curPerf>0].groupby(['Symbol']).curPerf.mean())
print(dfa[dfa.curPerf<0].groupby(['Symbol']).curPerf.mean())
```

```

dfa['Time'] = dfa.index
dfa['year'] = dfa.Time.apply(lambda x:x.year)
dfa
```

```

symbol = 'FXB'
lookback = 400


#dfa['perf3m'] =
#dfa.set_index(pd.to_datetime(dfa.Time))
dfa['perf90d'] = 0
dfa['perf30d'] = 0
dfa['perf180d'] = 0

dfa.loc[dfa.Symbol==symbol, 'perf90d'] = dfa[dfa.Symbol==symbol].rolling('90d', min_periods=5).curPerf.mean()
dfa.loc[dfa.Symbol==symbol, 'perf30d'] = dfa[dfa.Symbol==symbol].rolling('30d', min_periods=3).curPerf.mean()
dfa.loc[dfa.Symbol==symbol, 'perf180d'] = dfa[dfa.Symbol==symbol].rolling('180d', min_periods=5).curPerf.mean()

dfa[dfa.Symbol==symbol]['perf30d'].tail(lookback).plot(figsize=(12,8))
dfa[dfa.Symbol==symbol]['perf90d'].tail(lookback).plot(figsize=(12,8))
dfa[dfa.Symbol==symbol]['perf180d'].tail(lookback).plot(figsize=(12,8))
```

```
dfa[dfa.Symbol==symbol].rolling('30d', min_periods=5).curPerf.count()
```

```
#dfa['perf3m'] =
#dfa.set_index(pd.to_datetime(dfa.Time))
dfa['perf50p'] = 0
dfa['perf100p'] = 0
dfa['perf200p'] = 0

dfa.loc[dfa.Symbol==symbol, 'perf50p'] = dfa[dfa.Symbol==symbol].rolling(50, min_periods=5).curPerf.mean()
dfa.loc[dfa.Symbol==symbol, 'perf100p'] = dfa[dfa.Symbol==symbol].rolling(100, min_periods=5).curPerf.mean()
dfa.loc[dfa.Symbol==symbol, 'perf200p'] = dfa[dfa.Symbol==symbol].rolling(200, min_periods=5).curPerf.mean()

dfa[dfa.Symbol==symbol]['perf50p'].tail(lookback).plot(figsize=(12,8))
dfa[dfa.Symbol==symbol]['perf100p'].tail(lookback).plot(figsize=(12,8))
dfa[dfa.Symbol==symbol]['perf200p'].tail(lookback).plot(figsize=(12,8))
```

```
years = [y for y in dfa.year.unique()]
#print(dfa.year.unique())
res = {}
for y in  years:
    dfg = dfa[dfa.year==str(y)].groupby("Symbol").mean()['curPerf']
    #print(dfg.sort_values().tail(10))
    res[y] = dfg.rank().sort_values().tail(20)
res
```
