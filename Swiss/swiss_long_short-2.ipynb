```
# Start writing code here...
import pandas as pd
import numpy as np
import yfinance as yf
```

```
st = pd.read_csv("swiss_all.csv")
weekly = pd.DataFrame()    
for s in st.values:
    stock = s[0]
    tickers = yf.Ticker(stock)
    df = tickers.history(period = "10y")
    try:
        weekly[stock] = df.Close.resample('W-FRI').last()/df.Open.resample('W-FRI').first()-1
    except Exception as e:
        print(e)
        continue
    #print(stock)
    #break
weekly
```

```
weekly.fillna(-1).rank(axis=1)
```

```
div = int(st.count()/8)
bins = weekly.fillna(-1).rank(axis=1).apply(lambda x : x/div , axis=1).astype(int)
```

now wee need to figure out the average next value per bin see if there is any predictive power to that

```
for i in range(0, bins.max().max()):
    next = bins.applymap(lambda x:1 if x == i else np.nan)*bins.shift(-1)
    nextperf = bins.applymap(lambda x:1 if x == i else np.nan)*weekly.fillna(-1).shift(-1)
    print((i, next.mean().mean(), nextperf.mean().mean()))
    
```

looks like we need to short the worse performers and go long the best performers and get a really good ratio

```
def get_perf(short_bin, long_bin, weekly):
    bins = weekly.fillna(-1).rank(axis=1).apply(lambda x : x/10 , axis=1).astype(int)
    short = bins.applymap(lambda x:1 if x == short_bin else np.nan)*weekly.fillna(-1).shift(-1)
    longs = bins.applymap(lambda x:1 if x == long_bin else np.nan)*weekly.fillna(-1).shift(-1)
    res = (longs.mean(axis=1)-short.mean(axis=1)+1).cumprod()
    return res

def plot_perf(short_bin, long_bin, weekly, msg):
    res = get_perf(short_bin, long_bin, weekly)
    res.plot(figsize =(15, 10), legend=True, title=msg)
    return res
```

```
sr = []

for j in range(0, 9):
    for i in range(0, 9):
        if i < j:
            res = get_perf(j, i, weekly)
            if res[-2] > 4:# and res[-2] < 6:
                print(f"+{i} -{j}")
                res = plot_perf(j,i,weekly, msg=f"+{i} -{j}")
                sr.append([i, j, res[-2]/res.dropna().std()])
            # if res[-1] > 2:
            #     break
#print(f"+{i} -{j}")                
```
